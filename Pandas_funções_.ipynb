```python
# a função pd.DataFrame transoforma um conjunto de lista em um data frame, abaixo um exemplo de dataframe criado
data = {
    'Grupo': ['A', 'B', 'A', 'B', 'A', 'B'],
    'Valor': [10, 20, 30, 40, 50, 60]
}
data = pd.DataFrame(data)
```

```python
data_media = data.groupby('Grupo')['Valor'].mean()

display(data_media)
```

```python
# Funçao do python para realizar leitura de um arquivo csv

new_df = pd.read_csv('Electric_Vehicle_Population_Data.csv')

display(new_df)
```

```python
#função utilizada para excluir uma coluna do dataframe

new_df = new_df.drop('Clean Alternative Fuel Vehicle (CAFV) Eligibility',axis=1)

display(new_df)
```

```python
#função utilizada no pandas para renomar uma coluna do dataframe

new_df = new_df.rename(columns={'Postal Code':'ZIP Code'})

display(new_df)
```

```python
#Função no pandas para mesclar duas colunas do dataframe

new_df['Car model'] = new_df['Make'] + ' ' + new_df['Model']


display(new_df)
```
