```python
import pandas as pd
import numpy as np
import random
from datetime import datetime, timedelta

# Definindo os produtos e suas categorias correspondentes
produtos_categorias = {
    "bicicleta": "brinquedo",
    "aspirador de pó": "eletrodomésticos",
    "geladeira": "eletrodomésticos",
    "fogão": "eletrodomésticos",
    "cooktop": "eletrodomésticos",
    "sofa": "móveis",
    "lava e seca": "eletrodomésticos",
    "mesa": "móveis",
    "armario": "móveis",
    "guarda roupa": "móveis"
}

# Função para gerar uma data aleatória entre duas datas
def gerar_data_aleatoria(inicio, fim):
    return inicio + timedelta(days=random.randint(0, (fim - inicio).days))

# Função para gerar SKU único
def gerar_sku(produto):
    return f"{produto[:3].upper()}-{hash(produto) % 1000:03}"

# Período para as datas
inicio_data = datetime(2018, 1, 1)
fim_data = datetime(2022, 12, 31)

# Número de registros a serem gerados
num_registros = 1000

# Gerando os dados aleatórios
dados = []
for _ in range(num_registros):
    produto = random.choice(list(produtos_categorias.keys()))
    categoria = produtos_categorias[produto]
    quantidade = random.randint(1, 20)
    data_venda = gerar_data_aleatoria(inicio_data, fim_data)
    sku = gerar_sku(produto)
    dados.append([sku, produto, categoria, quantidade, data_venda])

# Criando um DataFrame com os dados gerados
df_vendas = pd.DataFrame(dados, columns=["SKU", "Produto", "Categoria", "Quantidade Vendida", "Data de Venda"])

# Exibindo os primeiros registros do DataFrame
print(df_vendas.head())
```
